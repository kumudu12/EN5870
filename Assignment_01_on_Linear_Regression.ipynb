2. Linear regression is a supervised learning technique used to model the relationship between a set of input features (X) and a target variable (y). Perform linear regression to learn parameters (weight vector) for the given data set.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

df_y = pd.read_csv('./y.csv')
df_y = df_y.values.reshape(-1)
print("Data shape: ", df_X.shape, df_y.shape)

#Split the data into training/testing sets
df_X_train = df_X[:-20]
df_X_test = df_X[-20:]

#Split the targets into training/testing sets
df_y_train = df_y[:-20]
df_y_test = df_y[-20:]

#Create linear regression model
regr = linear_model.LinearRegression()

#Train the model
regr.fit(df_X_train, df_y_train)

#Make predictions using the testing set
df_y_pred = regr.predict(df_X_test)

#The coeffcients
print("Coefficients:  ", regr.coef_)
#The mean squared error
print("Mean squared error: %.2f" %mean_squared_error(df_y_test, df_y_pred))
#The coefficient of determination:
print("Coefficient of determination: %.2f" % r2_score(df_y_test, df_y_pred))
```

```python
#Scatter plot
df_X_test=np.arange(0,len(df_X_test),1)
plt.scatter(df_X_test,df_y_test,color='blue')
plt.plot(df_X_test,df_y_pred,color='red',linewidth=2, label = 'Regression Line')
plt.xlabel('Input Feature (X)')
plt.ylabel('Target Variable (y)')
plt.title('Scatter Plot')
plt.legend()
```

12.  Next, we examine the effect of noisy measurements by introducing noise to the signal. Change the "noise_sigma" for values of 10âˆ’9, 10âˆ’3, 10âˆ’1, 1 and 10. Then, train the aforementioned models using the different noise levels. To observe how noise influences the performance of each model, plot the Mean Squared Error (MSE) between the estimated parameters (weight vector) and true parameters (given in "w.csv") for different noise levels. Comment on your results.


```python
noise_sigmas = [0.000000001, 0.001, 0.1, 1, 10] #noise levels

#Linear Model
MSE_linear_reg = []
print("MSEs for Linear Model")
for noise_sigma in noise_sigmas:
  df_y_noisy = df_y + noise_sigma * np.random.randn(len(df_y)) #add noise
  regr = linear_model.LinearRegression()
  regr.fit(df_X, df_y_noisy) #train the model
  MSE = mean_squared_error(df_w, regr.coef_) #calculate MSE between true W and learned w
  MSE_linear_reg.append(MSE)
print(MSE_linear_reg)

#Ridge Model
MSE_ridge_reg = []
print("MSEs for Ridge Model")
for noise_sigma in noise_sigmas:
  df_y_noisy = df_y + noise_sigma * np.random.randn(len(df_y))
  ridge_reg = Ridge(alpha=0.1)
  ridge_reg.fit(df_X, df_y_noisy)
  MSE = mean_squared_error(df_w, ridge_reg.coef_)
  MSE_ridge_reg.append(MSE)
print(MSE_ridge_reg)

#Lasso Model
MSE_lasso_reg = []
print("MSEs for Lasso Model")
for noise_sigma in noise_sigmas:
  df_y_noisy = df_y + noise_sigma * np.random.randn(len(df_y))
  lasso_reg = Lasso(alpha=0.1)
  lasso_reg.fit(df_X, df_y_noisy)
  MSE = mean_squared_error(df_w, lasso_reg.coef_)
  MSE_lasso_reg.append(MSE)
print(MSE_lasso_reg)
```

```python
#plot MSE between true parameters and learned parameters against different noise levels
plt.figure(figsize=(3, 3))
plt.plot(noise_sigmas, MSE_linear_reg, marker='*')
plt.title('Linear Model')
plt.xlabel('Noise Level')
plt.ylabel('Mean Squared Error (MSE)')
plt.grid(True)
plt.show()

plt.figure(figsize=(3, 3))
plt.plot(noise_sigmas, MSE_ridge_reg, marker='*')
plt.title('Ridge Model')
plt.xlabel('Noise Level')
plt.ylabel('Mean Squared Error (MSE)')
plt.grid(True)
plt.show()

plt.figure(figsize=(3, 3))
plt.plot(noise_sigmas, MSE_lasso_reg, marker='*')
plt.title('Lasso Model')
plt.xlabel('Noise Level')
plt.ylabel('Mean Squared Error (MSE)')
plt.grid(True)
plt.show()
```
